```python
filename = 'yes-minister.txt'
mytext = open(filename, encoding='utf-8').read()
```

```python
mytext
```

```python
from wordcloud import WordCloud
mycloud = WordCloud().generate(mytext)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(mycloud, interpolation='bilinear');
plt.axis('off');
```

```python
cnname = 'wordclound-cn.txt'
cntext = open(cnname, encoding='utf-8').read()
```

```python
import jieba
cutcn = ' '.join(jieba.cut(cntext))
```

```python
cn = WordCloud(font_path='simsun.ttf').generate(cutcn)
plt.imshow(cn, interpolation='bilinear');
plt.axis('off');
```
